# Data Visualization

# Total Stock Value by Category (Bar Chart)

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Category', y='Stock_Value', estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title("Total Stock Value by Category")
plt.ylabel("Stock Value (₹)")
plt.show()
```

# Stock Value vs. Sales Volume (Scatter Plot)

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Stock_Value', y='Sales_Volume', hue='Category')
plt.title("Stock Value vs. Sales Volume")
plt.show()
```

#  Stock Quantity Distribution (Histogram)

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Stock_Quantity'], bins=20, kde=True)
plt.title("Distribution of Stock Quantity")
plt.xlabel("Stock Quantity")
plt.ylabel("Frequency")
plt.show()
```

# Stock Age Analysis (Box Plot)

```python
df['Stock_Age'] = (pd.to_datetime('today') - pd.to_datetime(df['Last_Order_Date'])).dt.days

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Category', y='Stock_Age')
plt.xticks(rotation=45)
plt.title("Stock Age by Category")
plt.ylabel("Days since Last Order")
plt.show()
```

# Status Distribution (Pie Chart)

```python
plt.figure(figsize=(7, 7))
df['Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm')
plt.title("Product Status Distribution")
plt.ylabel("")
plt.show()
```

# Analysis of the Charts

**Total Stock Value by Category (Bar Chart):
"Fruits & Vegetables" have the highest total stock value, significantly exceeding other categories.
"Grains & Pulses" and "Oils & Fats" have relatively low stock values.
"Beverages", "Dairy" and "Seafood" have moderate stock values.**

Distribution of Stock Quantity (Histogram):
The distribution of stock quantity appears relatively uniform, with some variations.
There are products across the entire range of stock quantity.

**Stock Age by Category (Box Plot):
"Bakery" and "Seafood" categories have higher median stock ages (days since last order).
"Grains & Pulses" has a lower median stock age.
All categories show a wide range of stock ages.**

Product Status Distribution (Pie Chart):
The distribution of product status is relatively even, with "Discontinued", "Backordered", and "Active" products each accounting for roughly one-third of the total.**

**Stock Value vs. Sales Volume (Scatter Plot):
The majority of products have low stock values.
There's a wide range of sales volumes across all stock values.
There does not appear to be a strong direct correlation between stock value and sales volume.**

# Conclusion

**Inventory Imbalance: There's a significant imbalance in stock value across categories, with "Fruits & Vegetables" dominating.
High-Value Products: Specific products like "Banana" contribute significantly to the overall stock value.
Sales vs. Stock: Stock value doesn't necessarily correlate with sales volume, indicating potential inefficiencies in inventory management.
Stock Age Concerns: Certain categories, particularly "Bakery" and "Seafood," have older stock, which could lead to spoilage or obsolescence.
Product Status: The even distribution of product statuses suggests potential issues with product lifecycle management.**
